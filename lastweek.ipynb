```python
import pandas as pd
import statsmodels.formula.api as sm
```

```python
ds = pd.read_csv('dataset_fbw5.csv')
ds
```

```python
ds_dummy = pd.get_dummies(ds[['Payment', 'Type']])
ds_dummied = pd.concat([ds_dummy, ds], axis=1)
ds_dummied
```

```python
#Dependent Lifetime engaged 
# Independent status, video, link(LIKE?), paid and post month

model1 = sm.ols('Lifetime_Engaged_Users ~ Type_Status + Type_Video + Type_Link + Payment_Paid + Post_Month', data = ds_dummied).fit()
print(model1.summary())
```

```python
#Cook's D method to find the outliers and add to df
cooks_d = model1.get_influence().cooks_distance
n = len(ds_dummied)
ds_dummied['Outlier'] = cooks_d[0] > 4/n
#get a df with outliers and inspect and without
w_outlier_ds = ds_dummied[ds_dummied.Outlier == True]
no_outlier_ds = ds_dummied[ds_dummied.Outlier == False]

no_outlier_ds
```

```python
#without outliers
model2 = sm.ols('Lifetime_Engaged_Users ~ Type_Status + Type_Video + Type_Link + Payment_Paid + Post_Month', data = no_outlier_ds).fit()
print(f'Regression with outliers {model1.summary()}')
```

```python
print(f'Regression without outliers {model2.summary()}')
```

```python
import pandas as pd

df = pd.read_csv('insurance_missings.csv')
df.head()
```

```python
#Check how many missing values there are
df.isnull().sum()

#It seems BMI and Smoker have missing values
```

```python
#Prepare the data

#delete the missings
df2 = df.dropna()
df2.isnull().sum()

#create dummies
df2 = pd.get_dummies(df2)
df2.head()
```

```python
#Run the model

import statsmodels.formula.api as sm
model1 = sm.ols('charges~age+bmi+children+sex_male+smoker_yes', data = df2).fit()
print(model1.summary())
```

```python
#----------using imputation

#Create dummy variables for the missing data in the categorical variables (in this case smoker)

data = pd.get_dummies(df, dummy_na=True)

#use IterativeImputer to impute the remaining variables
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#impute the data
imputed_data = IterativeImputer().fit_transform(data)
#Save the data as a pandas dataframe and import the column names from the original dataset
imputed_data = pd.DataFrame(imputed_data, columns=data.columns)

#Run the new model
model2 = sm.ols('charges~age+bmi+children+sex_male+smoker_yes+smoker_nan', data = imputed_data).fit()
print(model2.summary())
```
